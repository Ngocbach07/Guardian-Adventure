### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
from urllib.request import urlretrieve
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chip = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
urlretrieve(chip, 'chipotle.tsv')
chipo=pd.read_csv('chipotle.tsv', sep = '\t')
chipo
```

### Step 4. How many products cost more than $10.00?

```python
cop = pd.DataFrame(chipo)
cop['item_price']=cop['item_price'].str.lstrip("$").astype(float)
newcop = cop[cop.item_price>10.00]
newcop.info()
```

### Step 5. What is the price of each item?
###### print a data frame with only three columns item_name choice_description and product_price

```python
chipo[['item_name', 'choice_description', 'item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
cop.sort_values('item_price', ascending = False, inplace= True)
cop['quantity'][3598]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
cop2=chipo.groupby('item_name')[['quantity']].sum()
cop2.loc['Veggie Salad Bowl']
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
cop3= chipo.set_index('item_name')
cop4= cop3.loc['Canned Soda']
ans = (cop4['quantity']>1).sum()
ans
```
